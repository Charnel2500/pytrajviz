```python
# demo.ipynb
# pyTrajViz – Interactive MD Trajectory Visualization Demo
# Run step by step

# STEP 1: Install dependencies (run only once)
#!pip install mdtraj plotly dash numpy scipy --quiet

# STEP 2: Import and load your trajectory
from pytrajviz.app import TrajVizApp

# CHANGE THESE PATHS TO YOUR FILES
traj_file = "eq4_200_201.xtc"   # your trajectory file
top_file  = "eq4.gro"           # your topology file (gro, pdb, etc.)

print("Loading trajectory...")
print(f"Trajectory: {traj_file}")
print(f"Topology:   {top_file}")

# Initialize the visualization app
app = TrajVizApp(traj_file, top_file)

# STEP 3: Launch the interactive web app (opens in a new browser tab)
print("Starting pyTrajViz...")
print("The app will open in your browser – click the link below!")

dash_app = app.build_app()
dash_app.run(port=8051, mode='external')

# After launching:
# → Use the slider to navigate through frames
# → Click + drag to rotate the 3D view
# → Scroll to zoom in/out
# → Hover over atoms to see coordinates
# → RMSD plot updates live with the current frame
```
